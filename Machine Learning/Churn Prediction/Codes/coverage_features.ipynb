```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Datasets/survey_client_with_churn.csv')
```

```python
df
```

```python
coverage = pd.read_excel('Datasets/Coverages.xlsx')
coverage = coverage.fillna(0)
```

```python
customers_list = list(df['Customer heading'].unique())
len(customers_list)
```

```python
coverage = coverage[coverage['Customer Heading'].isin(customers_list)]
```

```python
coverage
```

#### Covered this year by any representative (Yes = 1, No = 0)

```python
covered = coverage.groupby('Customer Heading').sum().drop('Current id_Representative', axis=1)
```

```python
df['Covered this year'] = df.apply(lambda row: covered.loc[row['Customer heading'], row.Year], axis=1)
```

#### How many customers this representative had in that year

```python
sellers = coverage.groupby('Current id_Representative').sum().drop('Customer Heading', axis=1)
```

```python
sellers.head()
```

```python
covered_by_seller = coverage.copy()

for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]:
    covered_by_seller[year] = covered_by_seller[year] * covered_by_seller['Current id_Representative']
    
covered_by_seller = covered_by_seller.groupby('Customer Heading').sum().drop('Current id_Representative', axis=1)

total_customers = covered_by_seller.copy()

def get_total_customers(year, x):
    
    try: 
        return sellers.loc[sellers[year].index==x, year].values[0]
    except:
        return 0

for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]:
    total_customers[year] = total_customers[year].map(lambda x: get_total_customers(year,x))
```

```python
df['Num customers of rep'] = df.apply(lambda row: total_customers.loc[row['Customer heading'], row.Year], axis=1)
```

#### The tenure of the representative

```python
sellers_tenure = np.cumsum(sellers > 0, axis=1)
```

```python
sellers_tenure.head()
```

```python
rep_tenure = covered_by_seller.copy()

def get_tenure(year, x):
    
    try: 
        return sellers_tenure.loc[sellers_tenure[year].index==x, year].values[0]
    except:
        return 0

for year in [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]:
    rep_tenure[year] = rep_tenure[year].map(lambda x: get_tenure(year,x))
```

```python
df['Tenure of rep'] = df.apply(lambda row: rep_tenure.loc[row['Customer heading'], row.Year], axis=1)
```

#### How many years a customer was covered in the past 3/4/5 years

```python
covered_3y = {2015: [], 2016: [], 2017: [], 2018: [], 2019:[]}
covered_4y = {2015: [], 2016: [], 2017: [], 2018: [], 2019:[]}

for year in [2015, 2016, 2017, 2018, 2019]:
    
    covered_3y[year] = covered[year-3] + covered[year-2] + covered[year-1]
    covered_4y[year] = covered[year-4] + covered[year-3] + covered[year-2] + covered[year-1]
    
covered_3y = pd.DataFrame(covered_3y)
covered_4y = pd.DataFrame(covered_4y)
```

```python
df['Covered in the past 3y'] = df.apply(lambda row: covered_3y.loc[row['Customer heading'], row.Year], axis=1)
df['Covered in the past 4y'] = df.apply(lambda row: covered_4y.loc[row['Customer heading'], row.Year], axis=1)
```

#### How many representatives the customer had in the past 3/4/5 years

```python
def get_unique_rep(row, years):
    
    year = row.Year
    customer = row['Customer heading']
    
    if years == 2:
        num = len(set(covered_by_seller.loc[customer, [year-1, year]]))
    elif years == 3:
        num = len(set(covered_by_seller.loc[customer, [year-2, year-1, year]]))
    elif years == 4:
        num = len(set(covered_by_seller.loc[customer, [year-3, year-2, year-1, year]]))
    elif years == 5:
        num = len(set(covered_by_seller.loc[customer, [year-4, year-3, year-2, year-1, year]]))
        
    return num
```

```python
df['Num of reps in 2y'] = df.apply(lambda row: get_unique_rep(row, 2), axis=1)
df['Num of reps in 3y'] = df.apply(lambda row: get_unique_rep(row, 3), axis=1)
df['Num of reps in 4y'] = df.apply(lambda row: get_unique_rep(row, 4), axis=1)
df['Num of reps in 5y'] = df.apply(lambda row: get_unique_rep(row, 5), axis=1)
```

#### How long (consecutive) the customer had the same representative

```python
def get_cons_years(row):
    
    year = row.Year
    customer = row['Customer heading']
    
    rep_list = covered_by_seller.loc[customer]
    
    num = 1
    current_rep = rep_list[year]
    
    while year > 2011:
        year = year - 1
        past_rep = rep_list[year]
    
        if (past_rep - current_rep) == 0:
            num = num + 1
        else:
            break
    
    return num
```

```python
df['Same rep in years'] = df.apply(lambda row: get_cons_years(row), axis=1)
```

#### How long the customer was covered by any representative

```python
def get_cons_years_any_rep(row):
    
    year = row.Year
    customer = row['Customer heading']
    
    rep_list = covered_by_seller.loc[customer]
    
    num = 1
    
    while year > 2011:
        year = year - 1
        past_rep = rep_list[year]
        
        if past_rep != 0:
            num = num + 1
        else:
            break
    
    return num
```

```python
df['Cons years covered'] = df.apply(lambda row: get_cons_years_any_rep(row), axis=1)
```

#### Weighted decaying coverage 

```python
def get_decaying_coverage(row):
    
    year = row.Year
    customer = row['Customer heading']
    
    rep_list = covered.loc[customer]
    
    num = 0
    denominator = 1

    while year >= 2011:
        
        num += rep_list[year] / denominator
        
        year = year - 1
        denominator += 1
    
    return num
```

```python
df['Decaying coverage'] = df.apply(lambda row: get_decaying_coverage(row), axis=1)
```

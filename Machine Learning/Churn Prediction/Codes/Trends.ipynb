```python
trends_spanish = survey_clean[survey_clean['Subject']=='spanish'].pivot_table(values='# copies', index=['Subject','Course'], 
                                     columns=['Year'], aggfunc=np.sum)

trends_spanish['R15_16'] = (100*(trends_spanish[2016])/(trends_spanish[2015]))
trends_spanish['R16_17'] = ((trends_spanish['R15_16'])*(trends_spanish[2017])/(trends_spanish[2016]))
trends_spanish['R17_18'] = ((trends_spanish['R16_17'])*(trends_spanish[2018])/(trends_spanish[2017]))
trends_spanish['R18_19'] = ((trends_spanish['R17_18'])*(trends_spanish[2019])/(trends_spanish[2018]))
trends_spanish
```

```python
trends_spanish = trends_spanish.drop([2015,2016,2017,2018,2019], axis = 1)
```

```python
trends_spanish = pd.DataFrame(trends_spanish)
trends_spanish
```

```python
import pandas as pd
import numpy as np
```

```python
math = pd.read_csv('predictions/mathematics.csv', index_col = 'Unnamed: 0')
natural = pd.read_csv('predictions/natural_sciences.csv', index_col = 'Unnamed: 0')
religion = pd.read_csv('predictions/religion.csv', index_col = 'Unnamed: 0')
social = pd.read_csv('predictions/social_sciences.csv', index_col = 'Unnamed: 0')
spanish = pd.read_csv('predictions/spanish.csv', index_col = 'Unnamed: 0')
music = pd.read_csv('predictions/music.csv', index_col = 'Unnamed: 0')
```

```python
test = pd.read_csv('testset.csv')
```

```python
math = math.groupby('Customer heading').predictions.agg(['count', 'median']).reset_index().rename(columns={'count': 'math_count', 'median': 'math_median'})
```

```python
natural = natural.groupby('Customer heading').predictions.agg(['count', 'median']).reset_index().rename(columns={'count': 'natural_count', 'median': 'natural_median'})
```

```python
religion = religion.groupby('Customer heading').predictions.agg(['count', 'median']).reset_index().rename(columns={'count': 'religion_count', 'median': 'religion_median'})
```

```python
social = social.groupby('Customer heading').predictions.agg(['count', 'median']).reset_index().rename(columns={'count': 'social_count', 'median': 'social_median'})
```

```python
spanish = spanish.groupby('Customer heading').predictions.agg(['count', 'median']).reset_index().rename(columns={'count': 'spanish_count', 'median': 'spanish_median'})
```

```python
music = music.groupby('Customer heading').predictions.agg(['count', 'median']).reset_index().rename(columns={'count': 'music_count', 'median': 'music_median'})
```

```python
math_natural = pd.merge(math, natural, how='outer')
ma_n_r = pd.merge(math_natural, religion, how='outer')
ma_n_r_s = pd.merge(ma_n_r, social, how='outer')
ma_n_r_s_sp = pd.merge(ma_n_r_s, spanish, how='outer')
df = pd.merge(ma_n_r_s_sp, music, how='outer')
```

```python
df = df.fillna(0)
```

```python
df['total teachers'] = df['math_count'] + df['religion_count'] + df['social_count'] + \
             df['spanish_count'] + df['music_count'] + df['natural_count']
```

```python
df.sort_values(['total teachers'], ascending=False)
```

```python
df['prob_sum'] = df['math_median'] + df['natural_median'] + df['religion_median'] + df['social_median'] + df['spanish_median'] + df['music_median']
```

```python
df['prob_mean'] = df['prob_sum'] / 6
```

```python
df['prob_per_teacher'] = df['total teachers'] / df['prob_sum']
```

```python
df.sort_values(['prob_mean', 'total teachers'], ascending=False)[:10]
```

```python
df.sort_values(['prob_mean', 'total teachers'], ascending=False)[:10].to_csv('top10_schools_to_target.csv', index=False)
```

```python
locations = pd.read_excel('Datasets/School locations.xlsx')
```

```python
locations_with_predictions = pd.merge(df, locations, on='Customer heading', how='left')
locations_with_predictions.to_csv('locations_with_predictions.csv', index=False)
```

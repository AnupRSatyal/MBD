```python
# How many people actually have a technical background? 

technical = dataset['Which best describes your undergraduate major? - Selected Choice'].value_counts() 
```

```python
technical = technical/technical.sum() * 100
```

```python
technical
```

```python
with plt.style.context('dark_background'):

# Plot
    plt.figure(figsize=(10,7))
    ordered_df = technical.sort_values(ascending=True)
    my_range=range(1,len(ordered_df.index)+1)
    plt.hlines(y=my_range, xmin=0, xmax=ordered_df, color='green',linestyles='dashed',alpha=0.5)
    plt.plot(ordered_df, my_range, "o", color='red' )

#     
    plt.title('Undergraduate Background (%)' ,fontsize=10)
    plt.tight_layout()
    plt.yticks(my_range, ordered_df.index)
    plt.xlabel('% of respondents')
    plt.tight_layout()
    plt.savefig('%respondents.png')
    plt.show()
    
```

```python
time = dataset[dataset['How many years of experience do you have in your current role?'] == '1-2'].loc[:,'During a typical data science project at work or school, approximately what proportion of your time is devoted to the following? (Answers must add up to 100%) - Cleaning data':
             'During a typical data science project at work or school, approximately what proportion of your time is devoted to the following? (Answers must add up to 100%) - Other']
```

```python
time.columns = ['Cleaning','Visualizing','Model Building','Model into Production','Communicating','Other']
```

```python
time = time.mean()
```

```python
time
```

```python
# Discount Leaders

with plt.style.context('dark_background'):

# Prepare Data
    df = time
    df.sort_values(inplace=True)

# Plot
    plt.figure(figsize=(10,7))
    plt.scatter(x=df.index, y=df, color='red', alpha=0.7, linewidth=20)
    
# Decoration
    plt.title('Time spent on specific tasks (%)', fontsize=10)
    plt.ylabel('Time spent by respondents')
    plt.xticks(df.index, df.index, rotation=60, horizontalalignment='right', fontsize=12)
    plt.tight_layout()
    plt.savefig('Time_spent.png')
```

```python
# The majority of Data Scientists are inexperienced

role_exp = dataset[['Select the title most similar to your current role (or most recent title if retired): - Selected Choice', 'How many years of experience do you have in your current role?']].dropna()
```

```python
role_exp.columns = ['Title','Experience']
```

```python
experience = role_exp[role_exp['Experience']== '1-2'].groupby('Title').count()
```

```python
experience  = (experience / experience.sum() * 100).sort_values('Experience')
```

```python
# Unit Leaders

with plt.style.context('dark_background'):

# Prepare Data
    df = experience
# Plot
    plt.figure(figsize=(10,7))
    sns.barplot(x=df.index,y=df.Experience, alpha=0.7, linewidth=20, color='red')

# Decoration
    plt.title('Job Role (% of Respondents)', fontsize=10)
    plt.ylabel('% of Respondents')
    plt.xticks(rotation=60, horizontalalignment='right', fontsize=12)
    plt.ylim((0,30))
    plt.xlabel('')
    plt.tight_layout()
    plt.savefig('experience')
```

```python
# Most famous libraries?

dataset['Of the choices that you selected in the previous question, which ML library have you used the most? - Selected Choice'].value_counts()
```

```python
language = dataset['What specific programming language do you use most often? - Selected Choice'].value_counts()
```

```python
language = language / language.sum()
```

```python
language
```

```python
import squarify 

with plt.style.context('dark_background'):

# Prepare Data
    df = language
    labels = df.index
    sizes = (df.values.tolist())
    colors = [plt.cm.Spectral(i/float(len(labels))) for i in range(len(labels))]

# Draw Plot
    plt.figure(figsize=(12,8), dpi= 80)
    squarify.plot(sizes=sizes, label=labels, color=colors, alpha=.8)

# Decorate
    plt.title('Most Used language (% of respondents)')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('language.png')
    plt.show()
```

```python
# What roles do people have?


role = dataset[dataset['How many years of experience do you have in your current role?'] == '1-2'].loc[:,'Select any activities that make up an important part of your role at work: (Select all that apply) - Selected Choice - Analyze and understand data to influence product or business decisions'
       :'Select any activities that make up an important part of your role at work: (Select all that apply) - Selected Choice - Other'].count()
```

```python
role.index = ['Analyse','Operational ML','Data Infrastructure','Innovative ML','ML Research','None','Other']
```

```python
role = role / role.sum()* 100
```

```python
with plt.style.context('dark_background'):

# Prepare Data
    df = role
    df.sort_values(inplace=True)

# Draw plot
    fig, ax = plt.subplots(figsize=(10,7), dpi= 80)
    ax.vlines(x=df.index, ymin=0, ymax=df, color='firebrick', alpha=0.7, linewidth=2)
    ax.scatter(x=df.index, y=df, s=75, color='firebrick', alpha=0.7)

# Title, Label, Ticks and Ylim
    ax.set_title('Role at work', fontdict={'size':10})
    ax.set_ylabel('% of respondents')
    ax.set_xticks(df.index)
    ax.set_xticklabels(df.index, rotation=60, fontdict={'horizontalalignment': 'right', 'size':12})
    ax.set_ylim(0,30)
    plt.tight_layout()
    plt.savefig('role')
```
